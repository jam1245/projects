```
# Create synthetic dataset (sine wave with no noise)
series = np.sin(0.1*np.arange(200)) 
#+np.random.randn(200)*01
```

```
# Plot the series
plt.plot(series)
plt.show()
```

```
# Build the dataset
# We are going to use T past values to predict the next value
# Cast X and Y as lists and then convert them to arrays
T = 10
X = []
Y = []

for t in range(len(series) - T):
  x = series[t:t+T]
  X.append(x)
  y = series[t+T]
  Y.append(Y)

X = np.array(X).reshape(-1, T)
Y = np.array(Y)
N = len(X)
print("X.shape", X.shape, "Y.shape", Y.shape)
```
